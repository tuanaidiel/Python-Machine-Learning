```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = iris.data
y = iris.target
X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
np.set_printoptions(suppress=True, precision=9) #precision=3
X_scaled
```

```python
# default value n_cluster is 8
# random (random), k-means++ (center)
# n_init is how many times to run after getting the cluster
# n_init higher the value will give you a good clusters but computation time gets slower
# max_iter is for each run

kmeans = KMeans(n_clusters=3, init='k-means++', n_init= 50, max_iter=100)
predictions = kmeans.fit_predict (X_scaled)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.scatter(X_scaled[predictions == 0, 0], X_scaled [predictions == 0, 1], c='red', label='Cluster 1')
plt.scatter(X_scaled[predictions == 1, 0], X_scaled [predictions == 1, 1], c='blue', label='Cluster 2')
plt.scatter(X_scaled[predictions == 2, 0], X_scaled [predictions == 2, 1], c='green', label='Cluster 3')


plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_ [:, 1], c='yellow', label='Centroid')
plt.legend()
```
